### Libs Importing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

### Dataset importing

```python
df = pd.read_csv('kc_house_data.csv')
df.head().T
```

## Dataset Exploring

***

```python
df.shape
```

```python
df.corr().round(4)
```

```python
## more relevant features comparing to 'price'
## correlation > 0.30

df = df[['id', 'date', 'price','bedrooms', 'bathrooms', 'sqft_living', 'view', 'grade', 
         'sqft_above', 'sqft_basement', 'sqft_living15']]

```

```python
df.describe()
```

```python
## date feature

df['date'] = df['date'].str.slice(start = 0, stop = 8) ## first 8 caracters (YYMMDD)
```

```python
df['date'] = df['date'].astype('datetime64') ## changing 'date' to datetime
```

```python
## getting month
df['month'] = df['date'].dt.month

## getting year
df['year'] = df['date'].dt.year

df
```

## Dataset Visualization

***

### Time Series 

```python
## mean price per date

avg_price_date = df.groupby('date').mean() ## date mean price 
avg_price_date = avg_price_date.reset_index() ## changing index

avg_price_year = df.groupby('year').mean() 
avg_price_year = avg_price_year.reset_index() 

avg_price_month = df.groupby('month').mean() 
avg_price_month = avg_price_month.reset_index() 
```

```python
fig3 = plt.figure(constrained_layout=True)
fig3 = plt.figure(figsize=(15,7))

gs = fig3.add_gridspec(2, 3) ## number of charts
graph1 = fig3.add_subplot(gs[0, :]) ##chart 1
graph1.set_title('AVG Price per year n month')
graph1 = plt.plot(avg_price_date['date'], avg_price_date['price'])


graph2 = fig3.add_subplot(gs[1, :-1])
graph2.set_title('AVG Price per month')
graph2 = plt.plot(avg_price_month['month'], avg_price_month['price'])
graph2 = plt.xlabel('month')


graph3 = fig3.add_subplot(gs[1:, -1])
graph3.set_title('AVG Price per year')
graph3 = plt.bar(avg_price_year['year'], avg_price_year['price'])
graph3 = plt.xlabel('year')
graph3 = plt.xticks(avg_price_year['year']) ## changing x axis 

fig3 = fig3
```

### Outliers

***

```python
##### Some basics statistics

meanprice = df['price'].mean()
minprice = df['price'].min()
maxprice = df['price'].max()
stdprice = df['price'].std()

print(f'Mean: {meanprice} | Min: {minprice} | Max: {maxprice}  | Std: {stdprice}')
```

```python
##### Boxplot

ax = sns.boxplot(data = df['price'], orient = 'v', width = 0.8)
ax.figure.set_size_inches(12,6)
ax.set_title('House Price')
```

```python
## Histogram

plt.figure(figsize=(10,5))
ax = sns.distplot(df["price"], bins = 100)

```

```python
### First outlier (10/2014)

first_date = datetime(2014,10, 1)
last_date = datetime(2014, 10, 31)

## getting only october, 2014
outlier1 = (df['date'] >= first_date) & (df['date'] <= last_date)

##creating a new column in original dataset
df['outlier1'] = outlier1

##getting only 'True'
outlier1 = df.query('outlier1')

## top 5 highest price
outlier1.nlargest(10, 'price')
```

```python
### Second outlier - 05/2015

first_date = datetime(2015, 5, 1)
last_date = datetime(2015, 5, 31)

## getting only june, 2015
outlier2 = (df['date'] >= first_date) & (df['date'] <= last_date)

## creating a new columns in original dataframe
df['outlier2'] = outlier2

## getting only 'True'
outlier2 = df.query('outlier2')

## top 5 highest price
outlier2.nlargest(10, 'price')
```

```python
## In order to explore the dataset further, columns 'outlier1' and 'outlier2' were 'dropped' 
##(not considered in variable df_false_outliers)
## to remember: these columns indicates dates 10/2014 and 05/2015 which contains some outliers prices

falseout1 = df[~df['outlier1']]
df_false_outlier = falseout1[~falseout1['outlier2']]

df_false_outlier.sort_values('price', ascending = False).head(10)
```
